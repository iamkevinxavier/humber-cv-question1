```python
# Importing necessary libraries
from skimage import io, color
from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Uploading image to colab
from google.colab import files
uploaded = files.upload()
```

```python
# Read the image
image = io.imread('rgb-2907595_1280.jpg')
io.imshow(image)
```

```python
# Check if the image is grayscale or RGB
if len(image.shape) == 2:  # Grayscale image
    rgb_image = color.gray2rgb(image)
else:  # RGB image
    rgb_image = image
```

```python
# Split the image into Red, Green, and Blue channels
red_channel = rgb_image.copy()
red_channel[:, :, 1] = 0  # Set green and blue channels to 0
red_channel[:, :, 2] = 0

green_channel = rgb_image.copy()
green_channel[:, :, 0] = 0  # Set red and blue channels to 0
green_channel[:, :, 2] = 0

blue_channel = rgb_image.copy()
blue_channel[:, :, 0] = 0  # Set red and green channels to 0
blue_channel[:, :, 1] = 0
```

```python
# Convert the image to grayscale for thresholding
gray_image = color.rgb2gray(rgb_image)

# Apply Otsu's thresholding
thresh = threshold_otsu(gray_image)
binary_image = gray_image > thresh
```

```python
print(thresh)
```

```python
# Inverted Thresholding
inverted_binary_image = gray_image <= thresh
```

```python
# Plot the images
fig, ax = plt.subplots(2, 4, figsize=(20, 10))

# RGB Image
ax[0, 0].imshow(rgb_image)
ax[0, 0].set_title('RGB Image')
ax[0, 0].axis('off')

# Red Channel
ax[0, 1].imshow(red_channel)
ax[0, 1].set_title('Red Channel')
ax[0, 1].axis('off')

# Green Channel
ax[0, 2].imshow(green_channel)
ax[0, 2].set_title('Green Channel')
ax[0, 2].axis('off')

# Blue Channel
ax[0, 3].imshow(blue_channel)
ax[0, 3].set_title('Blue Channel')
ax[0, 3].axis('off')

# Grayscale Image
ax[1, 0].imshow(gray_image, cmap='gray')
ax[1, 0].set_title('Grayscale Image')
ax[1, 0].axis('off')

# Binary Image (Thresholding)
ax[1, 1].imshow(binary_image, cmap='gray')
ax[1, 1].set_title('Binary Image (Thresholding)')
ax[1, 1].axis('off')

# Inverted Binary Image (Thresholding)
ax[1, 2].imshow(inverted_binary_image, cmap='gray')
ax[1, 2].set_title('Inverted Binary Image (Thresholding)')
ax[1, 2].axis('off')

# Empty plots for alignment
for i in range(3, 4):
    ax[1, i].axis('off')

plt.tight_layout()
plt.show()
```
